```python
import pandas as pd
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
df = pd.read_csv(url)
df
```

```python
df.dtypes
```

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
grouped = df.groupby('Response') 
```

```python
yes_grouped = grouped.get_group('Yes')
yes_grouped #filter on the yes response
```

```python
responded_yes_df = df[df['Response'].str.lower() == 'yes']
responded_yes_df
```

```python
yes_grouped_greater1000 = yes_grouped[yes_grouped['Total Claim Amount'] > 1000]
yes_grouped_greater1000 #filtered on yes et >1000 
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
average_claim_amount = responded_yes_df.groupby(['Policy Type', 'Gender'])['Total Claim Amount'].mean().reset_index()
average_claim_amount
#use .reset_index() to convert the index to columns and obtain a DataFrame
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# Group the DataFrame by 'State' and then count non-null 'Policy Type' entries for each group
state_policy_counts = df.groupby('State')['Customer'].count().reset_index()
state_policy_counts
```

```python
state_over500 = state_policy_counts[state_policy_counts['Customer'] > 500]
state_over500
```

4.Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
edu_gender_grouped = df.groupby(['Education', 'Gender'])
edu_gender_grouped['Customer Lifetime Value'].agg(['min','max','median'])
```
